## PhD Survey: Results well-being

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pingouin import cronbach_alpha
from scipy.stats import mannwhitneyu
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Define plotting helper function
def initialize_plot(ax):
    for pos in ('top', 'right', 'bottom', 'left'):
        ax.spines[pos].set_visible(False)

    ax.set_facecolor('ghostwhite')
    ax.grid(axis='y', color='lightgrey', linewidth=1)
    ax.set_axisbelow(True)
```

```python
# Set plotting variables
COLOR = "darkorange"
FIGURES = "figures/well-being/"

if not os.path.exists(FIGURES):
    os.makedirs(FIGURES)
```

### Read clean data

```python
# data = pd.read_csv("phd-survey-data-clean.csv", sep=";")
data = pd.read_csv("mock-data.csv", sep=";")
```

### DASS 21

```python
# Compute sum scores for subscales depression, anxiety, stress
data["DEPRESS_SUM"] = data[[f"DEPRESS_{x}" for x in range(1, 8)]].sum(axis=1)
data["ANXIETY_SUM"] = data[[f"ANXIETY_{x}" for x in range(1, 8)]].sum(axis=1)
data["STRESS_SUM"] = data[[f"STRESS_{x}" for x in range(1, 8)]].sum(axis=1)
```

```python
# Compute people above cut-off score for subscales
print("Screening positive")
print(f"For depression: {(data.DEPRESS_SUM > 10).to_list().count(True)}, {(data.DEPRESS_SUM > 10).to_list().count(True)/len(data)}")
print(f"For anxiety: {(data.ANXIETY_SUM > 6).to_list().count(True)}, {(data.ANXIETY_SUM > 6).to_list().count(True)/len(data)}")
print(f"For stress: {(data.STRESS_SUM > 10).to_list().count(True)}, {(data.STRESS_SUM > 10).to_list().count(True)/len(data)}")
```

```python
# Calculate Cronbach's alphas per subscale
print("CRONBACH'S ALPHAS")
print(f"Depression: {
cronbach_alpha(data[[f"DEPRESS_{i}" for i in range(1, 8)]])[0]:.2f}")
print(f"Anxiety: {
cronbach_alpha(data[[f"ANXIETY_{i}" for i in range(1, 8)]])[0]:.2f}")
print(f"Stress: {
cronbach_alpha(data[[f"STRESS_{i}" for i in range(1, 8)]])[0]:.2f}")
```

```python
# Plot depression, anxiety, stress (total, gender, nationality)
fig, ax = plt.subplots(3, 3, figsize=(16,16))

for i in range(3):
    for j in range(3):
        initialize_plot(ax[i][j])

for subscale, index in zip(("DEPRESS_SUM", "ANXIETY_SUM", "STRESS_SUM"), (0, 1, 2)):
    ax[index][0].hist(data[subscale], bins=np.arange(0, 22), color=COLOR)
    
    ax[index][1].hist(data[data.gender == "Man"][subscale], bins=np.arange(0, 22), color=COLOR, alpha=0.9, label="Men")
    ax[index][1].hist(data[data.gender == "Woman"][subscale], bins=np.arange(0, 22), color="royalblue", alpha=0.9, label="Women")
    ax[index][1].legend()
    
    ax[index][2].hist(data[data.nationality_new != "International"][subscale], bins=np.arange(0, 22), color=COLOR,
                  alpha=0.9, label='Austria/Germany')
    ax[index][2].hist(data[data.nationality_new == "International"][subscale], bins=np.arange(0, 22), color="royalblue",
                  alpha=0.9, label='International')
    ax[index][2].legend()

ax[0][0].set_title("Total")
ax[0][1].set_title("Men  vs. Women")
ax[0][2].set_title("International vs. Austria/Germany")

ax[0][1].set_xlabel("Depression score")
ax[1][1].set_xlabel("Anxiety score")
ax[2][1].set_xlabel("Stress score")

ax[0][0].set_ylabel("DEPRESSION\n\nFrequency")
ax[1][0].set_ylabel("ANXIETY\n\nFrequency")
ax[2][0].set_ylabel("STRESS\n\nFrequency")

# ax[0][0].text(12, 17, "n = 35 (17%)", color="royalblue", fontsize=10)
# ax[1][0].text(8, 24, "n = 39 (19%)", color="royalblue", fontsize=10)
# ax[2][0].text(12, 17, "n = 48 (23%)", color="royalblue", fontsize=10)

for line in range(3):  # indicate screening cut-offs
    ax[0][line].vlines(10, 0, 30, colors='royalblue', linestyle='--', linewidth=1)  # depression
    ax[1][line].vlines(6, 0, 50, colors='royalblue', linestyle='--', linewidth=1)  # anxiety
    ax[2][line].vlines(10, 0, 25, colors='royalblue', linestyle='--', linewidth=1)  # stress

fig.savefig(FIGURES + "depression-anxiety-stress.jpeg", dpi=300, bbox_inches='tight')
```
