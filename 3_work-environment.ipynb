## PhD Survey: Results work environment

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pingouin import cronbach_alpha
from scipy.stats import mannwhitneyu, t, spearmanr
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Define plotting helper function
def initialize_plot(ax):
    for pos in ('top', 'right', 'bottom', 'left'):
        ax.spines[pos].set_visible(False)

    ax.set_facecolor('ghostwhite')
    ax.grid(axis='y', color='lightgrey', linewidth=1)
    ax.set_axisbelow(True)
```

```python
# Set plotting variables
COLOR = "darkorange"
FIGURES = "figures/work-environment/"

if not os.path.exists(FIGURES):
    os.makedirs(FIGURES)
```

### Read clean data

```python
# data = pd.read_csv("phd-survey-data-clean.csv", sep=";")
data = pd.read_csv("mock-data.csv", sep=";")
```

## Helper functions for scale & CI computation

```python
# Define helper function to calculate subscale values
def scale_score(subscale):
    scores = []
    
    for indiv in range(len(data)):
        scores.append(data[[f"TEAM_{subscale}{x}" for x in range(1, 4)]].iloc[indiv].mean())

    return scores
```

```python
# Define helper function to calculate confidence interval based on t distribution
def compute_CI(d, cl=0.95):
    ci = t.interval(cl, df=len(d)-1, loc=np.mean(d), scale=np.std(d, ddof=1) / np.sqrt(len(d)))
    return ci
```

## Work environment general

```python
# Compute means and CIs for general items, separated for internationals vs. Austria/Germany
means_local = []
CIs_local = []
medians_local = []

means_int = []
CIs_int = []
medians_int = []

for var in ('nonrelevant_projects', 'external_collab', 'encourage_collab'):
    means_local.append(data[data.nationality_new != "International"][var].mean())
    means_int.append(data[data.nationality_new == "International"][var].mean())

    CIs_local.append(compute_CI(data[data.nationality_new != "International"][var]))
    CIs_int.append(compute_CI(data[data.nationality_new == "International"][var]))

    medians_local.append(data[data.nationality_new != "International"][var].median())
    medians_int.append(data[data.nationality_new == "International"][var].median())
```

```python
# Plot setup
fig, ax = plt.subplots(figsize=(7, 4))

plt.axvspan(0, 2.5, 0.33, 1, facecolor='lightcoral', alpha=0.2)
plt.axvspan(2.5, 5, 0.33, 1, facecolor='lightgreen', alpha=0.2)
plt.axvspan(2.5, 5, 0, 0.33, facecolor='lightcoral', alpha=0.2)
plt.axvspan(0, 2.5, 0, 0.33, facecolor='lightgreen', alpha=0.2)

# Plot horizontal error bars (mean ± std)
ax.errorbar(means_local, np.arange(len(means_local)*2, step=2), xerr=[[mean - ci[0] for mean, ci in zip(means_local, CIs_local)],
                                                                      [ci[1] - mean for mean, ci in zip(means_local, CIs_local)]], 
            fmt='o', color='orangered', ecolor=COLOR, capsize=3, linewidth=1, label="Austria (n=136) & Germany (n=18)")

ax.scatter(medians_local, [0, 2, 4], marker='^', color='orangered', label='Medians')
ax.scatter(medians_int, [1, 3, 5], marker='^', color='orangered')

ax.errorbar(means_int, np.arange(1, len(means_int)*2+1, step=2), xerr=[[mean - ci[0] for mean, ci in zip(means_int, CIs_int)],
                                                                      [ci[1] - mean for mean, ci in zip(means_int, CIs_int)]], 
            fmt='x', color='orangered', ecolor=COLOR, capsize=3, linewidth=1, label="Internationals (n=50)")

ax.set_yticks(np.arange(0.5, 5, step=2))
ax.set_yticklabels(["Nonrelevant projects", 
                    "Collaborating externally",
                    "Collaborations encouraged"])
ax.set_xticklabels(["Strongly disargee", 
                    "Disagree",
                    "Slightly disagree",
                    "Slightly agree",
                    "Agree",
                    "Strongly agree"],
                    rotation=45, ha='center')

ax.text(5.2, 3.35, "Error bars show 95% CIs.\nDifferences tested with Mann Whitney U.", color="black", fontsize=10)

ax.set_title("Work environment")
ax.set_xlim(0, 5)
ax.grid(True, axis='x', linestyle='--', alpha=0.5)
ax.legend(bbox_to_anchor=(1.01, 0.76))

fig.savefig(FIGURES + "work-environment.jpeg", dpi=300, bbox_inches='tight')
```

## Team atmosphere: Collaborative and/or competitive

```python
# Compute scale scores for collabortive and competitive atmosphere
data["TEAM_COLLAB"] = scale_score("collab")
data["TEAM_COMP"] = scale_score("comp")
```

```python
print("CRONBACH'S ALPHAS")
print(f"Collaboration: {
cronbach_alpha(data[[f"TEAM_collab{x}" for x in range (1, 4)]])[0]:.2f}")
print(f"Competition: {
cronbach_alpha(data[[f"TEAM_comp{x}" for x in range (1, 4)]])[0]:.2f}")
print(f"Correlation between the two scales: {spearmanr(data["TEAM_COLLAB"], data["TEAM_COMP"])[0]:.2f}")
```

```python
# Plot collaborative vs. competitive atmosphere
counts = data[["TEAM_COLLAB", "TEAM_COMP"]].value_counts().reset_index(name="count")

fig, ax = plt.subplots(figsize=(7, 7))

plt.axvspan(-0.25, 2.5, 0.5, 1, facecolor='lightgreen', alpha=0.2)
plt.axvspan(2.5, 5.25, 0, 0.5, facecolor='lightcoral', alpha=0.2)
plt.axvspan(-0.25, 2.5, 0, 0.5, facecolor='ghostwhite')
plt.axvspan(2.5, 5.25, 0.5, 1, facecolor='ghostwhite')

ax.scatter(counts.TEAM_COMP, counts.TEAM_COLLAB, s=counts["count"]*25, color=COLOR)

ax.set_ylim(-0.25, 5.25)
ax.set_xlim(-0.25, 5.25)

ax.vlines(2.5, -0.25, 5.25, colors='darkgrey', linestyle='--', linewidth=1)
ax.hlines(2.5, -0.25, 5.25, colors='darkgrey', linestyle='--', linewidth=1)

ax.set_xlabel("Competitive")
ax.set_ylabel("Collaborative")

ax.text(1.15, 3.75, "Collaborative\nNon-competitive", horizontalalignment="center", color="black")
ax.text(1.15, 1.1, "Non-collaborative\nNon-competitive", horizontalalignment="center", color="black")
ax.text(3.85, 3.75, "Collaborative\nCompetitive", horizontalalignment="center", color="black")
ax.text(3.85, 1.1, "Non-collaborative\nCompetitive", horizontalalignment="center", color="black")

ax.set_title("Work environment")

fig.savefig(FIGURES + "collab-comp.jpeg", dpi=300, bbox_inches='tight')
```

## Integration into institutions

```python
# Compute means and CIs for different institutions, separated for men vs. women, internationals vs. Austria/Germany
means_total = []
means_men = []
means_women = []
means_ints = []
means_locals = []

CIs_men = []
CIs_women = []
CIs_ints = []
CIs_locals = []

for var in ["INTEGRATE_uni", "INTEGRATE_phd", "INTEGRATE_docschool", "INTEGRATE_institute", "INTEGRATE_group"]:
    means_total.append(data[var].mean())
    means_men.append(data[data.gender == 'Man'][var].mean())
    means_women.append(data[data.gender == 'Woman'][var].mean())
    means_ints.append(data[data.nationality_new == "International"][var].mean())
    means_locals.append(data[data.nationality_new != "International"][var].mean())

    CIs_men.append(compute_CI(data[data.gender == 'Man'][var]))
    CIs_women.append(compute_CI(data[data.gender == 'Woman'][var]))
    CIs_ints.append(compute_CI(data[data.nationality_new == "International"][var]))
    CIs_locals.append(compute_CI(data[data.nationality_new != "International"][var]))
```

```python
# Plot integration into institutions
fig, ax = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Total
ax[0].barh(np.arange(0.225, 5.225), means_total, height=0.4, color=COLOR)
ax[0].set_yticks(np.arange(0.225, 5.225))
ax[0].set_yticklabels(["University", "PhD community", "Docschool", "Institute", "Research group"])
ax[0].set_title("Total")

# Men vs. women
ax[1].barh(np.arange(5), means_women, height=0.4, color="royalblue", label="Women")
ax[1].barh(np.arange(5)+0.45, means_men, height=0.4, color=COLOR, label="Men")

ax[1].errorbar(means_women, np.arange(len(means_women)), xerr=[[mean - ci[0] for mean, ci in zip(means_women, CIs_women)],
                                                               [ci[1] - mean for mean, ci in zip(means_women, CIs_women)]], 
               fmt=' ', ecolor='black', capsize=2, linewidth=0.75)
ax[1].errorbar(means_men, np.arange(0.45, len(means_men)+0.45), xerr=[[mean - ci[0] for mean, ci in zip(means_men, CIs_men)],
                                                                      [ci[1] - mean for mean, ci in zip(means_men, CIs_men)]], 
               fmt=' ', ecolor='black', capsize=2, linewidth=0.75)
ax[1].legend()
ax[1].set_title("Men vs. Women")

# Internationals vs. Austria/Germany
ax[2].barh(np.arange(5), means_ints, height=0.4, color="royalblue", label="Internationals")
ax[2].barh(np.arange(5)+0.45, means_locals, height=0.4, color=COLOR, label="Austria & Germany")

ax[2].errorbar(means_ints, np.arange(len(means_ints)), xerr=[[mean - ci[0] for mean, ci in zip(means_ints, CIs_ints)],
                                                              [ci[1] - mean for mean, ci in zip(means_ints, CIs_ints)]], 
               fmt=' ', ecolor='black', capsize=2, linewidth=0.75)
ax[2].errorbar(means_locals, np.arange(0.45, len(means_locals)+0.45), xerr=[[mean - ci[0] for mean, ci in zip(means_locals, CIs_locals)],
                                                                            [ci[1] - mean for mean, ci in zip(means_locals, CIs_locals)]], 
               fmt=' ', ecolor='black', capsize=2, linewidth=0.75)
ax[2].legend()
ax[2].set_title("International vs. Austria/Germany")

# Set some plot parameters
for a in (0, 1, 2):
    initialize_plot(ax[a])
    ax[a].grid(True, axis='x', linestyle='--', alpha=0.5)
    ax[a].grid(False, axis='y')
    ax[a].set_xticks(np.arange(0, 6, step=1))
    ax[a].set_xticklabels(["Strongly disargee", 
                           "Disagree",
                           "Slightly disagree",
                           "Slightly agree",
                           "Agree",
                           "Strongly agree"],
                           rotation=45, ha='right')

fig.savefig(FIGURES + "integration.jpeg", dpi=300, bbox_inches='tight')
```
