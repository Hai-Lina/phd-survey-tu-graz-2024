## PhD Survey: Results supervision

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pingouin import cronbach_alpha
from scipy.stats import mannwhitneyu
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Define plotting helper function
def initialize_plot(ax):
    for pos in ('top', 'right', 'bottom', 'left'):
        ax.spines[pos].set_visible(False)

    ax.set_facecolor('ghostwhite')
    ax.grid(axis='y', color='lightgrey', linewidth=1)
    ax.set_axisbelow(True)
```

```python
# Set plotting variables
COLOR = "darkorange"
FIGURES = "figures/supervisor-relationship/"

if not os.path.exists(FIGURES):
    os.makedirs(FIGURES)
```

### Read clean data

```python
# data = pd.read_csv("phd-survey-data-clean.csv", sep=";")
data = pd.read_csv("mock-data.csv", sep=";")
```

### Supervisor relationship (scale)

```python
# Define helper function to calculate subscale values
def scale_score(subscale, length):
    scores = []
    
    for indiv in range(len(data)):
        scores.append(data[[f"SUPERVISOR_{subscale}{x}" for x in range(1, length+1)]].iloc[indiv].mean())

    return scores
```

```python
# Calculate subscale values
data["SUPERVISOR_SAFE"] = scale_score('safe', 5)
data["SUPERVISOR_COMMIT"] = scale_score('commit', 4)
data["SUPERVISOR_REFLECT"] = scale_score('reflect', 3)
data["SUPERVISOR_MODEL"] = scale_score('model', 3)
data["SUPERVISOR_FEEDBACK"] = scale_score('feedback', 2)

SUPERVISOR_TOTAL = []

for indiv in range(len(data)):
    SUPERVISOR_TOTAL.append(data[["SUPERVISOR_SAFE", "SUPERVISOR_COMMIT", "SUPERVISOR_REFLECT",
                                  "SUPERVISOR_MODEL", "SUPERVISOR_FEEDBACK"]].iloc[indiv].mean())

data["SUPERVISOR_TOTAL"] = SUPERVISOR_TOTAL
```

```python
# Calculate Cronbach's alphas per subscale and in total
print("CRONBACH'S ALPHAS")
print(f"Safe base: {
cronbach_alpha(data[["SUPERVISOR_safe1", "SUPERVISOR_safe2", "SUPERVISOR_safe3", "SUPERVISOR_safe4", "SUPERVISOR_safe5"]])[0]:.2f}")
print(f"Commitment: {
cronbach_alpha(data[["SUPERVISOR_commit1", "SUPERVISOR_commit2", "SUPERVISOR_commit3", "SUPERVISOR_commit4"]])[0]:.2f}")
print(f"Reflective education: {
cronbach_alpha(data[["SUPERVISOR_reflect1", "SUPERVISOR_reflect2", "SUPERVISOR_reflect3"]])[0]:.2f}")
print(f"Role model: {
cronbach_alpha(data[["SUPERVISOR_model1", "SUPERVISOR_model2", "SUPERVISOR_model3"]])[0]:.2f}")
print(f"Formative feedback: {
cronbach_alpha(data[["SUPERVISOR_feedback1", "SUPERVISOR_feedback2"]])[0]:.2f}")
print(f"Total: {
cronbach_alpha(data[[x for x in data.columns if 'SUPERVISOR' in x][:-5]])[0]:.2f}")
```

```python
# Correlations (Spearman) between subscales
data[["SUPERVISOR_SAFE", "SUPERVISOR_COMMIT", "SUPERVISOR_REFLECT",
      "SUPERVISOR_MODEL", "SUPERVISOR_FEEDBACK"]].corr(method='spearman')
```

```python
# Plot total, men vs woman, international vs Austria/Germany for supervisor relationship
fig, ax = plt.subplots(1, 3, figsize=(15,5))
initialize_plot(ax[0])
initialize_plot(ax[1])
initialize_plot(ax[2])

ax[0].hist(data.SUPERVISOR_TOTAL, bins=np.arange(0, 5.5, 0.5), color=COLOR)
# ax[0].vlines(2.5, 0, 52, colors='royalblue', linestyle='--', linewidth=1)
# ax[0].text(1, 34, "n=35 (17%)", color="royalblue", fontsize=10)
ax[0].set_title("Total")

ax[1].hist(data[data.gender == "Man"].SUPERVISOR_TOTAL, bins=np.arange(0, 5.5, 0.5),
           color=COLOR, alpha=0.9, label='Men')
ax[1].hist(data[data.gender == "Woman"].SUPERVISOR_TOTAL, bins=np.arange(0, 5.5, 0.5),
           color='royalblue', alpha=0.9, label='Women')
ax[1].legend()
ax[1].set_title("Men vs. Women")

ax[2].hist(data[data.nationality_new != "International"].SUPERVISOR_TOTAL, bins=np.arange(0, 5.5, 0.5),
           color=COLOR, alpha=0.9, label='Austria/Germany')
ax[2].hist(data[data.nationality_new == "International"].SUPERVISOR_TOTAL, bins=np.arange(0, 5.5, 0.5),
           color='royalblue', alpha=0.9, label='International')
ax[2].legend()
ax[2].set_title("International vs. Austria/Germany")

for a in (ax[0], ax[1], ax[2]):
    a.set_xticks(np.arange(0, 5.5, step=1))
    a.set_xticklabels(["Strongly disargee", 
                        "Disagree",
                        "Slightly disagree",
                        "Slightly agree",
                        "Agree",
                        "Strongly agree"],
                        rotation=45, ha='right')

fig.savefig(FIGURES + "supervisor-relationship.jpeg", dpi=300, bbox_inches='tight')
```

```python
# Test differences supervisor relationship in MEN VS. WOMEN
mannwhitneyu(data[data.gender == "Man"].SUPERVISOR_TOTAL,
             data[data.gender == "Woman"].SUPERVISOR_TOTAL)
```

```python
# Test differences supervisor relationship in INTERNATIONALS VS. AUSTRIA/GERMANY
mannwhitneyu(data[data.nationality_new != "International"].SUPERVISOR_TOTAL,
             data[data.nationality_new == "International"].SUPERVISOR_TOTAL)
```

```python
# Calculate means per doc school
ds_super = dict()

for ds in set(data.docschool_stud):
    ds_super[ds] = data[data.docschool_stud == ds].SUPERVISOR_TOTAL.median()

ds_super["Techno-Economics"] = 0
ds_super["Geosciences"] = 0

del ds_super["Not say"]

ds_super_sorted = {k: v for k, v in sorted(ds_super.items(), key=lambda item: item[1])}
```

```python
# Plot means per doc school
fig, ax = plt.subplots()
initialize_plot(ax)

ax.barh(ds_super_sorted.keys(), ds_super_sorted.values(), color=COLOR)
ax.text(0.15, 0.85, "omitted for data protection", color="darkorange", fontsize=10)
ax.text(0.15, -0.15, "omitted for data protection", color="darkorange", fontsize=10)
ax.grid(True, axis='x', linestyle='--', alpha=0.5)
ax.grid(False, axis='y')
ax.set_title("Supervisor relationship")

fig.savefig(FIGURES + "supervisor-per-docschool.jpeg", dpi=300, bbox_inches='tight')
```
