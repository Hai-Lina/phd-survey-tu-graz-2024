## PhD Survey: Results teaching and working conditions

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

from scipy.stats import pointbiserialr, mannwhitneyu
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Define plotting helper function
def initialize_plot(ax):
    for pos in ('top', 'right', 'bottom', 'left'):
        ax.spines[pos].set_visible(False)

    ax.set_facecolor('ghostwhite')
    ax.grid(axis='y', color='lightgrey', linewidth=1)
    ax.set_axisbelow(True)
```

```python
# Set plotting variables
COLOR = "darkorange"
FIGURES = "figures/teaching-research/"

if not os.path.exists(FIGURES):
    os.makedirs(FIGURES)
```

### Read clean data

```python
# data = pd.read_csv("phd-survey-data-clean.csv", sep=";")
data = pd.read_csv("mock-data.csv", sep=";")
```

### Teaching hours & lectures

```python
# Compute "giving lectures - yes, no" for UnivAss and ProjectAss and store in dict
lectures_univ = dict()
lectures_project = dict()

for answer in ('Yes', 'No'):
    lectures_univ[answer] = data[data.employ_status == "UnivAss"].lectures_YN.to_list().count(answer)
    lectures_project[answer] = data[(data.employ_status == "ProjectAss TU") | (data.employ_status == "ProjectAss Center")
                                   ].lectures_YN.to_list().count(answer)
```

```python
# Plot teaching hours & lectures
fig, ax = plt.subplots(1, 3, figsize=(12,4))
initialize_plot(ax[0])
initialize_plot(ax[1])
initialize_plot(ax[2])

# Teaching hours total
ax[0].hist(data.teaching_hours, bins=np.arange(0, 41, step=5), color=COLOR)
ax[0].set_xlabel("Hours")
ax[0].set_title("Total")

# Teaching hours per employment status
ax[1].hist(data[data.employ_status == "UnivAss"].teaching_hours, bins=np.arange(0, 41, step=5), color=COLOR, alpha=0.9, label="UnivAss")
ax[1].hist(data[(data.employ_status == "ProjectAss TU") | (data.employ_status == "ProjectAss Center")
               ].teaching_hours, bins=np.arange(0, 41, step=5), color="royalblue", alpha=0.8, label="ProjectAss")
ax[1].legend()
ax[1].set_xlabel("Hours")
ax[1].set_title("UnivAss vs. ProjectAss")

# Lectures per employment status
ax[2].bar(lectures_univ.keys(), lectures_univ.values(), color=COLOR, label="UnivAss")
ax[2].bar(lectures_project.keys(), lectures_project.values(), bottom=lectures_univ.values(), color='royalblue', alpha=0.8, label="ProjectAss")
ax[2].legend()
ax[2].set_title("Giving lectures")

fig.savefig(FIGURES + "working-hours-lectures.jpeg", dpi=300, bbox_inches='tight')
```

### Relationship lectures <> teaching hours (UnivAss)

```python
# Compute correlation between giving lectures & working hours for UnivAss
pointbiserialr(data[data.employ_status == "UnivAss"].lectures_YN.replace({"Yes": 1, "No": 0}),
               data[data.employ_status == "UnivAss"].teaching_hours)
```

```python
no_data = data[(data.employ_status == "UnivAss") & (data.lectures_YN == "No")].teaching_hours
yes_data = data[(data.employ_status == "UnivAss") & (data.lectures_YN == "Yes")].teaching_hours

# Plot correlation giving lectures & working hours for UnivAss
fig, ax = plt.subplots()
initialize_plot(ax)

boxprops = dict(facecolor='royalblue', alpha=0.7)
medianprops = dict(color=COLOR, linewidth=2)

ax.boxplot([no_data, yes_data], positions=[0, 1], widths=0.3, boxprops=boxprops, medianprops=medianprops, patch_artist=True)

# ax.text(0.4, 12, "r = .38", color="black", fontsize=10)

ax.set_xticklabels(['No', 'Yes'])
ax.set_xlabel("Giving lectures")
ax.set_ylabel('Working hours')
ax.set_title('Lectures <> Working hours (UnivAss)')

fig.savefig(FIGURES + "working-hours-CORR-lectures.jpeg", dpi=300, bbox_inches='tight')
```

### Hours in contract

```python
# Compute hours in contract per employment status
contract_univ = dict()
contract_project = dict()

for answer in ('<20', '21-30', '31-40'):
    contract_univ[answer] = data[data.employ_status == "UnivAss"].hours_contract.to_list().count(answer)
    contract_project[answer] = data[(data.employ_status == "ProjectAss TU") | (data.employ_status == "ProjectAss Center")
                                   ].hours_contract.to_list().count(answer)
```

```python
# Plot hours in contract per employment status
fig, ax = plt.subplots()
initialize_plot(ax)

ax.bar(contract_univ.keys(), contract_univ.values(), color=COLOR, label="UnivAss")
ax.bar(contract_project.keys(), contract_project.values(), bottom=contract_univ.values(),
          color='royalblue', alpha=0.8, label="ProjectAss")
ax.legend()

ax.set_xlabel("Hours in  contract")

fig.savefig(FIGURES + "hours-in-contract.jpeg", dpi=300, bbox_inches='tight')
```

### Actual working hours

```python
# Plot working hours
fig, ax = plt.subplots(1, 4, figsize=(16,4))
initialize_plot(ax[0])
initialize_plot(ax[1])
initialize_plot(ax[2])
initialize_plot(ax[3])

# Working hours total
ax[0].hist(data.working_hours, bins=np.arange(0, 61, step=5), color=COLOR)
ax[0].set_xticks(np.arange(0, 61, step=10))
ax[0].set_xlabel("Hours")
ax[0].set_title("Total")

# Working hours per employment status
ax[1].hist(data[(data.employ_status == "ProjectAss TU") | (data.employ_status == "ProjectAss Center")
               ].working_hours, bins=np.arange(0, 61, step=5), color="royalblue", alpha=0.9, label="ProjectAss")
ax[1].hist(data[data.employ_status == "UnivAss"].working_hours, bins=np.arange(0, 61, step=5), color=COLOR, alpha=0.9, label="UnivAss")
ax[1].legend()
ax[1].set_xlabel("Hours")
ax[1].set_title("UnivAss vs. ProjectAss")

# Working hours per gender
ax[2].hist(data[(data.gender == "Man")].working_hours, bins=np.arange(0, 61, step=5), color=COLOR, alpha=0.9, label="Men")
ax[2].hist(data[(data.gender == "Woman")].working_hours, bins=np.arange(0, 61, step=5), color="royalblue", alpha=0.9, label="Women")
ax[2].legend()
ax[2].set_xlabel("Hours")
ax[2].set_title("Men vs. Women")

# Working hours per nationality
ax[3].hist(data[data.nationality_new != "International"].working_hours, bins=np.arange(0, 61, step=5),
           color=COLOR, alpha=0.9, label="Austria/Germany")
ax[3].hist(data[data.nationality_new == "International"].working_hours, bins=np.arange(0, 61, step=5),
           color="royalblue", alpha=0.9, label="International")
ax[3].legend()
ax[3].set_xlabel("Hours")
ax[3].set_title("International vs. Austria/Germany")

fig.savefig(FIGURES + "working-hours.jpeg", dpi=300, bbox_inches='tight')
```

### Overtime

```python
# Create new variable capturing max hours in contract
hours_contract_new = []

for i in data.hours_contract:
    if i == "<20":
        hours_contract_new.append(20)
    if i == "21-30":
        hours_contract_new.append(30)
    if i == "31-40":
        hours_contract_new.append(40)
    if i == "No employee":
        hours_contract_new.append("No employee")

data["hours_contract_new"] = hours_contract_new
```

```python
# Compute overtime (actual working hours - max hours in contract)
df_overtime = data[["working_hours", "hours_contract_new", "employ_status", "gender", "nationality_new"]]
df_overtime = df_overtime[df_overtime.hours_contract_new != "No employee"]
df_overtime["overtime"] = df_overtime.working_hours - df_overtime.hours_contract_new
```

```python
# Plot overtime
fig, ax = plt.subplots(1, 4, figsize=(16,4))
initialize_plot(ax[0])
initialize_plot(ax[1])
initialize_plot(ax[2])
initialize_plot(ax[3])

# Overtime total
ax[0].hist(df_overtime.overtime, bins=np.arange(0, 41, step=5), color=COLOR)
ax[0].set_xticks(np.arange(0, 41, step=10))
ax[0].set_xlabel("Hours")
ax[0].set_title("Total")

# Overtime per employment status
ax[1].hist(df_overtime[(df_overtime.employ_status == "ProjectAss TU") | (df_overtime.employ_status == "ProjectAss Center")
               ].overtime, bins=np.arange(0, 41, step=5), color="royalblue", alpha=0.9, label="ProjectAss")
ax[1].hist(df_overtime[df_overtime.employ_status == "UnivAss"].overtime, bins=np.arange(0, 41, step=5),
           color=COLOR, alpha=0.9, label="UnivAss")
ax[1].legend()
ax[1].set_xlabel("Hours")
ax[1].set_title("UnivAss vs. ProjectAss")

# Overtime per gender
ax[2].hist(df_overtime[(df_overtime.gender == "Man")].overtime, bins=np.arange(0, 41, step=5), color=COLOR, alpha=0.9, label="Men")
ax[2].hist(df_overtime[(df_overtime.gender == "Woman")].overtime, bins=np.arange(0, 41, step=5),
           color="royalblue", alpha=0.9, label="Women")
ax[2].legend()
ax[2].set_xlabel("Hours")
ax[2].set_title("Men vs. Women")

# Overtime per nationality
ax[3].hist(df_overtime[df_overtime.nationality_new != "International"].overtime, bins=np.arange(0, 41, step=5),
           color=COLOR, alpha=0.9, label="Austria/Germany")
ax[3].hist(df_overtime[df_overtime.nationality_new == "International"].overtime, bins=np.arange(0, 41, step=5),
           color="royalblue", alpha=0.9, label="International")
ax[3].legend()
ax[3].set_xlabel("Hours")
ax[3].set_title("International vs. Austria/Germany")

fig.savefig(FIGURES + "overtime.jpeg", dpi=300, bbox_inches='tight')
```

### Compromising circumstances

```python
# Helper function to calculate percentage points of different compromising circumstances per group
def compute_circum(df):
    CIRCUM = dict()
    
    CIRCUM["Social relationships"] = (df.CIRCUM_social.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Compromised health"] = (df.CIRCUM_health.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Living conditions"] = (df.CIRCUM_life.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Financial situation"] = (df.CIRCUM_financial.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Child care"] = (df.CIRCUM_carechild.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Care work (other)"] = (df.CIRCUM_careother.to_list().count("Yes") / len(df)) * 100
    
    CIRCUM["Other"] = (len(df.CIRCUM_free_text.dropna()) / len(df)) * 100
    CIRCUM["None of the above"] = (df.CIRCUM_none.to_list().count("Yes") / len(df)) * 100
    CIRCUM["Prefer not to say"] = (df.CIRCUM_notsay.to_list().count("Yes") / len(df)) * 100
    
    return CIRCUM
```

```python
circum_total = compute_circum(data)
circum_men = compute_circum(data[data.gender == "Man"])
circum_women = compute_circum(data[data.gender == "Woman"])
circum_local = compute_circum(data[data.nationality_new != "International"])
circum_int = compute_circum(data[data.nationality_new == "International"])
```

```python
# Plot circumstances compromising PhD progress
fig, ax = plt.subplots(1, 3, figsize=(12, 6), sharey=True)
initialize_plot(ax[0])
initialize_plot(ax[1])
initialize_plot(ax[2])

# Total
ax[0].barh(np.arange(0.225, 9.225), circum_total.values(), color=COLOR, height=0.4)
ax[0].set_yticks(np.arange(0.225, 9.225))
ax[0].set_yticklabels(circum_total.keys())
ax[0].invert_yaxis()
ax[0].set_title("Total")

# Men vs. women
ax[1].barh(np.arange(9), circum_men.values(), height=0.4, color=COLOR, label="Men")
ax[1].barh(np.arange(9)+0.45, circum_women.values(), height=0.4, color="royalblue", label="Women")
ax[1].legend()
ax[1].set_title("Men vs. Women")

# Internationals vs. Austria/Germany
ax[2].barh(np.arange(9), circum_local.values(), height=0.4, color=COLOR, label="Austria & Germany")
ax[2].barh(np.arange(9)+0.45, circum_int.values(), height=0.4, color="royalblue", label="Internationals")
ax[2].legend()
ax[2].set_title("International vs. Austria/Germany")

for a in (0, 1, 2):
    ax[a].grid(True, axis='x', linestyle='--', alpha=0.5)
    ax[a].grid(False, axis='y')
    ax[a].xaxis.set_major_formatter(PercentFormatter(decimals=0))

fig.savefig(FIGURES + "circumstances.jpeg", dpi=300, bbox_inches='tight')
```
